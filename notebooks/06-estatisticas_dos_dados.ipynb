```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ratings = pd.read_csv('../dados/ratings.csv')
movies = pd.read_csv('../dados/movies.csv')
```

```python
movie1_name = movies.query('movieId == 1')["title"].values[0]
movie2_name = movies.query('movieId == 2')["title"].values[0]

movie1_ratings = ratings.query('movieId == 1')["rating"]
movie2_ratings = ratings.query('movieId == 2')["rating"]

movie1_ratings_mean = np.mean(movie1_ratings)
movie2_ratings_mean = np.mean(movie2_ratings)

movie1_ratings__standard_deviation = np.std(movie1_ratings)
movie2_ratings_standard_deviation = np.std(movie2_ratings)

print(f"A média de avaliações do filme {movie1_name} é {movie1_ratings_mean:.2f}")
print(f"A média de avaliações do filme {movie2_name} é {movie2_ratings_mean:.2f}")

print(f"O desvio padrão das avaliações do filme Toy Story é {movie1_ratings__standard_deviation:.2f}")
print(f"O desvio padrão das avaliações do filme Jumanji é {movie2_ratings_standard_deviation:.2f}")
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot([movie1_ratings, movie2_ratings], labels=[movie1_name, movie2_name])
plt.title(f"Distribuição de avaliações dos filmes {movie1_name} e {movie2_name}")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(
    data=ratings.query('movieId in [1, 2]'),
    x='movieId',
    y='rating',
).set(
    xticklabels=[movie1_name, movie2_name],
    xlabel='', 
    ylabel='Nota'
)
plt.title(f"Distribuição de avaliações dos filmes {movie1_name} e {movie2_name}")
plt.grid(axis='y')
plt.show()
```

```python
# Há filmes com poucas avaliações, o que pode enviesar sua média devido a pouca amostragem
movies_with_one_rating = ratings.groupby('movieId').count().query("rating < 100").shape[0]
print(f"Há {movies_with_one_rating} filmes com menos de 100 avaliações.")
```
