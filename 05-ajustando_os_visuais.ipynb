```python
import pandas as pd
import seaborn as sns # https://seaborn.pydata.org/
import matplotlib.pyplot as plt
```

```python
# Exibindo idiomans em ordem crescente de quantidade
# https://seaborn.pydata.org/tutorial/color_palettes.html

tmdb = pd.read_csv('./dados/tmdb_5000_movies.csv')

movies_that_are_not_in_english = tmdb.query("original_language != 'en'")
languages_different_than_english = movies_that_are_not_in_english["original_language"].value_counts(normalize=True)

plt.figure(figsize=(15, 5)) # tamanho dao próximo gráfico (largura x altura)
sns.countplot(
    data=movies_that_are_not_in_english, 
    order=languages_different_than_english.index, # ordem de exibicao dos elementos no eixo x
    x="original_language", 
    hue="original_language", # elementos que devem ser considerados na paleta de cores
    stat="percent", # tipo de estatística a ser exibida
    palette="rocket", # paleta de cores
    hue_order=languages_different_than_english.index, # ordem dos elementos para a paleta de cores
    )

plt.title("Porcentagem dos idimas originais dos filmes que não são em inglês")
plt.xlabel("Idioma original")
plt.ylabel("Total (%)")
plt.show()
```
