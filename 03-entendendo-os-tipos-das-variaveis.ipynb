```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importando a base de dados
tmdb = pd.read_csv('./dados/tmdb_5000_movies.csv')
tmdb.head(3)
```

```python
# Criação do gráfico de receitas dos filmes
sns.displot(tmdb["revenue"])
plt.title("Distribuição das receita dos filmes")
plt.ylabel("Número de filmes")
plt.xlabel("Receita")
plt.show()
```

```python
# Criação do gráfico de orçamento dos filmes
sns.displot(tmdb["budget"])
plt.title("Distribuição das orçamento dos filmes")
plt.ylabel("Número de filmes")
plt.xlabel("Orçamento")
plt.show()
```

```python
# Exibindo o tipo das variáveis/colunas
tmdb.info()
```

```python
# Exibindo estatísticas descritivas das variáveis/colunas
tmdb.describe()
```

```python
# Exibindo filmes cujo orçamento está registrado na base de dados
movies_with_revenue = tmdb.query("revenue > 0")
sns.displot(movies_with_revenue["revenue"])
plt.title("Distribuição das receita dos filmes (com receita registrada)")
plt.ylabel("Número de filmes")
plt.xlabel("Receita")
plt.show()
```
